```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
x2 = np.linspace(-5.0, 5.0, 100)
y2 = np.sqrt(10**2 - x2**2)
y2 = np.hstack([y2, -y2])
x2 = np.hstack([x2, -x2])
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)
y1 = np.hstack([y1, -y1])
x1 = np.hstack([x1, -x1])
```

```python
plt.scatter(y2,x2)
plt.scatter(y1,x1)
```

```python
x1_new = x1
y1_new = y1*y1
x2_new = x2
y2_new = y2*y2
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.scatter(y2_new, x2_new)
ax.scatter(y1_new, x1_new)
```
