```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x1 = np.array([1,2,3,9,10,11,12])
y1 = np.array([1,2,3,9,10,11,12])
x2 = np.array([4,5,6,7,8])
y2 = np.array([4,5,6,7,8])
```

```python
fig, axs = plt.subplots(1, 2)

plt.title("Polynomial kernel")

axs[0].set_title("before")
axs[0].scatter(x2, y2)
axs[0].scatter(x1, y1)

y1_new = y1**8
y2_new = y2**8

axs[1].set_title("after")
axs[1].scatter(x2, y2_new)
axs[1].scatter(x1, y1_new)


plt.show()
```

```python
circle_pts = [(math.sin(x/100),math.cos(x/100)) for x in range(-int(100*math.pi),+int(100*math.pi),1)]
circle_pts2 = [(2*math.sin(x/100),2*math.cos(x/100)) for x in range(-int(100*math.pi),+int(100*math.pi),1)]
```

```python
fig = plt.figure()
plt.title("Polynomial kernel")
ax1 = fig.add_subplot(1, 2, 1, projection="3d")
ax1.set_title("before")
ax1.scatter(np.array([x for x, y in circle_pts]), np.array([y for x, y in circle_pts]))
ax1.scatter(
    np.array([x for x, y in circle_pts2]), np.array([y for x, y in circle_pts2])
)

# # y1_new = y1**8
# # y2_new = y2**8

circle_pts_new = [(x**2, y**2, x * y) for x, y in circle_pts]
circle_pts2_new = [(x**2, y**2, x * y) for x, y in circle_pts2]


ax1 = fig.add_subplot(1, 2, 2, projection="3d")
ax1.set_title("after")
ax1.scatter(
    np.array([x for x, y, z in circle_pts_new]),np.array([y for x, y, z in circle_pts_new]), np.array([z for x, y, z in circle_pts_new])
)
ax1.scatter(
    np.array([x for x, y, z in circle_pts_new]),np.array([y for x, y, z in circle_pts2_new]), np.array([z for x, y, z in circle_pts2_new])
)


plt.show()
```
